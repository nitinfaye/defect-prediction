## Checking for threshold

```python
col_names = df.columns
corr_matrix = df.corr()
cat_cols = ['uniq_Op','uniq_Opnd', 'total_Op','total_Opnd','branchCount']
corr_threshold = 0.8                                                

print("****Correlation****")
for c in col_names:
    for r in col_names:
        if (r < c):
            if (c not in cat_cols) and (c is not 'defect') and (r not in cat_cols) and (r is not 'defect') :
                if abs(corr_matrix[r][c]) > corr_threshold:
                    print(c +" : "+ r+" : "+ str(corr_matrix[r][c]))
```

This shows the highest and the lowest correlations as per the correlation threshold i.e. 0.8 in our case, it shows only those correlations which are 0.8 or greater than that, but in no particular sequence.

### Outlier treatment

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
import numpy as np
for column in df.iloc[:, 1:10].columns:
    lr,ur=remove_outlier(df[column])
    df[column]=np.where(df[column]>ur,ur,df[column])
    df[column]=np.where(df[column]<lr,lr,df[column])
```

This is to treat the outlier which shows the lower range follows from the the Inter-Quartile Range which is the from the first quarter and the upper range is from the upper quarter which is 75% and the Inter-Quartile Range. 

After the function ran, we can see that this range from the lower and upper range is giving us a value that cancels the outliers which are within the range and if they are outside the range, the outliers will not be treated well.

```python
plt.figure(figsize=(10,10))
df.iloc[:, 0:7].boxplot(vert=0)
```

We can see here, **loc** has outliers which are outside the range of the function that we have specified and hence, the outliers are not treated in this case

# Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
NB_model = GaussianNB()
NB_model.fit(x_train, y_train)
```

```python
y_train_predict = NB_model.predict(x_train)
model_score = NB_model.score(x_train, y_train)
print(model_score)
print(metrics.confusion_matrix(y_train, y_train_predict))
print(metrics.classification_report(y_train, y_train_predict))
```
